```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Zadanie 6.1

```python
#1
plt.figure(figsize=(10, 5), dpi=300, num=1)
#3
plt.plot([1,3,5,9], label='data 3')
#5
plt.plot([2,4,6,10], color='red', marker='s', linewidth=0,
         label='data 2')
#6
plt.plot([3,5,7,11], color='green', marker='s', markersize=12,
         linestyle='--', linewidth=1, label='data 3')
#8
plt.title('figure 1', fontsize=16, fontweight='bold')
#9
plt.xlabel('numbers X', fontsize=12, color='grey')
plt.ylabel('numbers Y', fontsize=12, color='grey')
#10
plt.legend(loc=4, fontsize=16)
#11
plt.figtext(0.2, 0.75, 'important \ndescription',
            horizontalalignment='center',
            bbox={'facecolor': 'yellow', 'alpha':0.5, 'pad':5})
#12
plt.xlim([-1, 4])
plt.ylim([-1, 13])
#13
ax = plt.gca()
ax.yaxis.grid(which='major')
ax.set_yticks(np.arange(0, 13, 0.5), minor=True)
ax.yaxis.grid(which='minor', linestyle='--')
#7
#plt.savefig('figure1.png')
#4
plt.show()
#2
plt.close(1)
```

### Zadanie 6.2

```python
dataset1 = pd.read_csv('student-mat.csv')
plt.figure(figsize=(10, 5), dpi=300, num=2)
plt.plot('age', data=dataset1, marker='*', linewidth=0)
plt.show()
plt.close(2)
```

```python
dataset1 = dataset1.sort_values(by=['age'])
plt.figure(figsize=(10, 5), dpi=300, num=3)
plt.plot(np.arange(len(dataset1['age'])), 'age', data=dataset1,
         marker='*', linewidth=0)
plt.show()
plt.close(3)
```
