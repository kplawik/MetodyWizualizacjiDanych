### Zadanie 6.5

```python
dataset2 = pd.read_csv('student-mat.csv')
dataset3 = pd.DataFrame(np.array([[dataset2[dataset2['sex'] == 'M']['sex'].count(), 
                                   dataset2[dataset2['sex'] == 'M']['age'].mean(),
                                   dataset2[dataset2['sex'] == 'M']['age'].min(),
                                   dataset2[dataset2['sex'] == 'M']['age'].max()], [
                                   dataset2[dataset2['sex'] == 'F']['sex'].count(), 
                                   dataset2[dataset2['sex'] == 'F']['age'].mean(),
                                   dataset2[dataset2['sex'] == 'F']['age'].min(),
                                   dataset2[dataset2['sex'] == 'F']['age'].max()]]),
    index=['M', 'F'],
    columns=['number', 'avg_age', 'min_age', 'max_age'])
dataset3.index.name = 'sex'
dataset3['avg_age'] = dataset3['avg_age'].round(2)
dataset3
```

```python
plt.scatter(dataset3.index, dataset3['avg_age'], marker='*', label="AVG")
plt.scatter(dataset3.index, dataset3['min_age'], marker='_', label="MIN")
plt.scatter(dataset3.index, dataset3['max_age'], marker='+', label="MAX")
plt.legend(loc=10)
plt.show()
```

### Zadanie 6.6

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figheight(5)
fig.set_figwidth(10)
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
xLabel = np.arange(-7, 7, 1)
ax1.plot(x, np.sin(x))
ax1.set_title('sin')
ax2.plot(x, np.cos(x))
ax2.set_title('cos')
ax2.set_xticks(xLabel)
ax2.set_xticklabels(xLabel, rotation=90)
fig.suptitle('Trigonometry')
fig.tight_layout()
plt.show()
```

### Zadanie 6.7

```python

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.set_figheight(10)
fig.set_figwidth(10)
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
xLabel = np.arange(-7, 7, 1)

ax1.plot(x, np.sin(x))
ax1.set_title('sin')
ax2.plot(x, np.cos(x))
ax2.set_title('cos')
ax2.set_xticks(xLabel)
ax2.set_xticklabels(xLabel, rotation=90)
ax3.plot(x, np.sin(x)) 
ax3.plot(x, np.cos(x))
ax3.set_title('sin and cos')
ax3.set_xticks(xLabel)
ax3.set_xlim([-np.pi, np.pi])
ax4.plot(x, np.sin(x) + np.cos(x))
ax4.set_title('sin + cos')
fig.suptitle('Trigonometry')
fig.tight_layout()
plt.show()
```

### Zadanie 6.16

```python
pivot_table = pd.pivot_table(dataset2, index='age', columns='sex', values='school', aggfunc='count', fill_value=0)
students = pd.DataFrame(pivot_table['F'] + pivot_table['M'], columns=['students'])

print(pivot_table)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_figheight(4)
fig.set_figwidth(10)
ax1.pie(pivot_table['M'], explode=(0, 0, 0, 0, 0, 0.5, 0.5, 0.5), labels=pivot_table.index,
        normalize=True, autopct='%.0f%%')
ax1.set_title('M')
ax2.pie(pivot_table['F'], explode=(0, 0, 0, 0, 0, 0.5, 0.5, 0.5), labels=pivot_table.index,
        normalize=True, autopct='%.0f%%')
ax2.set_title('F')
ax3.pie(students['students'], explode=(0, 0, 0, 0, 0, 0.5, 0.5, 0.5), labels=students.index,
        normalize=True, autopct='%.0f%%')
ax3.set_title('M + F')
fig.suptitle('Age')
plt.tight_layout()
plt.show()
```
